```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Criando o DataFrame
data = {
    'Entidade': [
        'IPT', 'IPT', 'IPT', 'Cruz Vermelha', 'Cruz Vermelha', 'Cruz Vermelha',
        'Corpo de Bombeiros', 'Corpo de Bombeiros', 'Corpo de Bombeiros', 'Corpo de Bombeiros',
        'SMADS', 'SMADS', 'CPAS', 'CPAS', 'CCOI', 'CCOI', 'DVISAM', 'DVISAM', 'DVISAM', 'DVISAM',
        'COVISA', 'COVISA', 'COVISA', 'COVISA', 'Comunicação Social', 'Comunicação Social', 
        'Comunicação Social', 'Comunicação Social', 'Comunicação Social', 'Comunicação Social',
        'Polícia', 'Samu', 'Defesa Civil Município de São Paulo', 'Defesa Civil Município de São Paulo',
        'Defesa Civil Estadual', 'Defesa Civil Estadual', 'Defesa Civil Estadual', 'Defesa Civil Estadual',
        'Defesa Civil Estadual', 'Defesa Civil Estadual', 'GCM', 'GCM', 'GCM', 'Academia', 'Academia'
    ],
    'Meio de Comunicação': [
        'Chamada Telefônica', 'WhatsApp/Telegram', 'Pessoalmente', 'Chamada Telefônica', 
        'WhatsApp/Telegram', 'Base de Apoio Local', 'Chamada Telefônica', 'WhatsApp/Telegram', 
        'Pessoalmente', 'Rádio Comunicador', 'Chamada Telefônica', 'WhatsApp/Telegram', 
        'Chamada Telefônica', 'WhatsApp/Telegram', 'Chamada Telefônica', 'WhatsApp/Telegram', 
        'Chamada Telefônica', 'WhatsApp/Telegram', 'Pessoalmente', 'E-mail', 'Chamada Telefônica', 
        'WhatsApp/Telegram', 'Pessoalmente', 'E-mail', 'Chamada Telefônica', 'WhatsApp/Telegram', 
        'Pessoalmente', 'Redes Sociais', 'Rádio Comunicador', 'Comunicação por Satélite ("Starlink")', 
        'Chamada Telefônica', 'Chamada Telefônica', 'Chamada Telefônica', 'WhatsApp/Telegram', 
        'Chamada Telefônica', 'WhatsApp/Telegram', 'Pessoalmente', 'Rádio Comunicador', 
        'Redes Sociais', 'Comunicação por Satélite ("Starlink")', 'Chamada Telefônica', 
        'WhatsApp/Telegram', 'Rádio Comunicador', 'WhatsApp/Telegram', 'Outras Redes Sociais'
    ]
}

df = pd.DataFrame(data)
```

```python
# Contando a frequência de cada meio de comunicação
comunicacao_freq = df['Meio de Comunicação'].value_counts()

# Exibindo os resultados em formato de tabela
print(comunicacao_freq)

# Visualização gráfica da distribuição dos meios de comunicação
plt.figure(figsize=(10, 6))
sns.barplot(x=comunicacao_freq.values, y=comunicacao_freq.index, palette='Blues_d')
plt.title('Distribuição dos Meios de Comunicação Mais Utilizados', fontsize=16)
plt.xlabel('Frequência de Uso', fontsize=14)
plt.ylabel('Meio de Comunicação', fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# Contar o número de meios de comunicação por entidade
entidade_comunicacao = df.groupby('Entidade')['Meio de Comunicação'].nunique()

# Exibir a tabela
print(entidade_comunicacao)

# Visualização gráfica
plt.figure(figsize=(10, 6))
entidade_comunicacao.plot(kind='bar', color='skyblue')
plt.title('Número de Meios de Comunicação Utilizados por Entidade')
plt.xlabel('Entidade')
plt.ylabel('Número de Meios de Comunicação')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
